```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [11,22,33,22]
```

```python
plt.plot(x,y)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
fig, ax = plt.subplots()
ax.set(title = "Example", xlabel = "Example x label", ylabel = "Example y label")
ax.plot(x,y)
```
